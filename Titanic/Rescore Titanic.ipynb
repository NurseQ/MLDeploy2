### Relationship between vars_with_na and Survived

```python
def analyze_na_vars(data,var):
    # make a copy of the dataframe so original data is not overwritten
    data = data.copy()
    
     # make a interimg variable that indicates 1 if the value/observation is 
    # missing and 0 otherwise
    data[var] = np.where(data[var].isnull(),1,0)
    data[var].value_counts().plot
```

### Check if other variables are categorical

```python
for var in num_vars:
    print(data[var].value_counts())
    print()
```

```python
discrete_vars=['SibSp', 'Parch']

for var in discrete_vars:
    sns.catplot(x=var, hue='Survived', data=data, kind='count')
```

```python
def analyze_vars(data,var):
    # make a copy of the dataframe so original data is not overwritten
    data = data.copy()
    
     # make a interimg variable that indicates 1 if the value/observation is 
    # missing and 0 otherwise
    data[var] = np.where(data[var]==0,0,1)
    print(data[var].value_counts(normalize=True))
    print(data.groupby(var)['Survived'].value_counts(normalize=True))
    sns.catplot(x=var, hue='Survived', data=data, kind='count', height=3)
    
```

```python
for var in discrete_vars:
    analyze_vars(data,var)  
```

The above plot shows people with or without SibSp/Parch on board against there survival. It appears that those without SibSp/Parch have similar outcome where only about half survived. Those with SibSp/Parch, it appears that there is nearly the same chance of survival and with the case of those with Parch, it even shows a slightly higher chance of surivival.
