### Feature Selection

Select a subset of the most predictive features. There is an element of randomness in Lasso Regression, hence it needs to set the seed.


```python
# Model fitting and feature selection
# alpha is the penalty, bigger penalty means less features will be selected

sel_ = SelectFromModel(Lasso(alpha=0.001, random_state=0))

# train the model and select features
sel_.fit(X_train, y_train)
```

```python
sel_.get_support().sum()
```

```python
selected_feats = X_train.columns[(sel_.get_support())]

print('total features: {}'.format((X_train.shape[1])))
print('selected features: {}'.format(len(selected_feats)))
print('selected with coefficients shrank to zero: {}'.format(np.sum(sel_.estimator_.coef_==0)))
```
